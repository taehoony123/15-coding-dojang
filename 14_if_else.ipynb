```python
# else 사용하기
x = 5

if x == 10:
    print("10")
else :
    print("10이 아니다.")
```

```python
# else 사용하기
x = 5

if x == 10:
    print("10")
else :
    print("x에 들어있는 숫자는")
    print(f"{x}이다.")
```

```python
a, b, c, d = map(int, input().split())
print(a, b, c, d)

if 0 <= a <= 100 and 0 <= b <= 100 and 0 <= c <= 100 and 0 <= d <= 100:
    total = a + b + c + d
    avg = total / 4

    if avg >= 80:
        print("합격")
    else:
        print("불합격")
else:
    print("0에서 100 사이의 숫자만 입력하세요")
```
