```python
# 딕셔너리 만들기
lux = {
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}
print(type(lux))
print(lux)
```

```python
x = {
    100 : "hundered",
    False : 0,
    3.5 : [3.5, 3.5]
}
print(type(x))
print(x)
```

```python
# 딕셔너리 키 로는 리스트, 딕셔너리 사용 불가
y = {
    [10,20] : 100
}
print(y)
print(type(y))
```

```python
z = {
    {"a" : 10} : 100
}

print(z)
print(type(z))
```

```python
z = {
    100 : {"a" : 10} 
}

print(z)
print(type(z))
```

```python
# 빈 딕셔너리 만들기

x = {}

y = dict()

print(x,y)
print(type(x),type(y))
print(bool(x),bool(y))
```

```python
# dict 함수로 딕셔너리 만들기
lux1 = dict(health = 400 , mane = 334, melee = 50, armor = 10.72)
print(lux1)
```

```python
lux2 = dict(zip(["health", "mane", "melee", "armor"],[400,334,50,10.72]))
print(lux2)
print(type(lux2))
```

```python
lux3 = dict([("health",490),("mane",334),("melee",50), ("armor",10.72)])
print(lux3)
print(type(lux3))
```

```python
# 딕셔너리의 키에 값 할당하기
lux = {
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}

lux["health"]
```

```python
lux["health"] = 2037
print(lux)
```

```python
lux["mana_ragen"] = 3.28
print(lux)
```

```python
# 딕셔너리에 키가 있는지 확인하기
print(lux)
print("health" in lux )
print("attach_speed" in lux)
```

```python
# 딕셔너리에 키의 개수 구하기
print(len(lux))
```
