## 함수 안에서 전역 변수 변경 global

```python
x = 10 # 전역변수 
def foo():
    x = 20 # x는 foo의 지역변수
    print(x) # foo의 지역변수 출력

foo()
print(x)
```

```python
x = 10 # 전역변수
def foo():
    global x #전역변수  x를 사용하겠다고 설정
    x = 20 # x는 전역 변수
    print(x) # 전역변수 출력

foo()
print(x)
```

```python
# 전역 변수 x가 없는 상태
def foo():
    global x    # x를 전역 변수로 만듦
    x = 20      # x는 전역 변수
    print(x)    # 전역 변수 출력
 
foo()
print(x)        # 전역 변수 출력
```

## 지역 변수의 범위

바깥쪽 함수의 지역 변수는 그 안에 속한 모든 함수에서 접근할 수 있습니다.

```python
def print_hello():
    hello = 'Hello, world!'
    def print_message():
        print(hello)    # 바깥쪽 함수의 지역 변수를 사용
```

```python
def A():
    x = 10 # 바깥쪽 지역변수
    def B():
        x = 20 # 안쪽 지역변수

    B() 
    print(x) # 바깥쪽 

A()
print(x)
```

```python
def A():
    x = 10 # 바깥쪽 지역변수
    def B():
        nonlocal x
        x = 20 # 안쪽 지역변수

    B() 
    print(x) # 바깥쪽 

A()
print(x)
```

## nonlocal이 변수를 찾는 순서

가까운 함수부터 지역 변수를 찾고, 지역 변수가 없으면 계속 바깥쪽으로 나가서 찾습니다.<br>
def C()에서 nonlocal을 하게 되면 def B() 에서 찾고 def A() 순으로 찾는다 

즉, 가까운 함수부터 지역변수를 찾고 지역변수가 없으면 바깥쪽으로 나가서 찾음

```python
def A():
    x = 10
    y = 100
    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x = x + 30
            y = y + 300
            print(x)
            print(y)
        C()
    B()
 
A()
```

시험 안나옴

```python
def A():
    x = 10
    y = 100
    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x = x + 30
            y = y + 300
            print('1',x)
            print('2',y)
        print('3',x)
        C()
    
    print('4',x)
    print('5',y)
    B()
    
print('6',)
A()
```

##  global로 전역 변수 사용하기

```python
x = 1
def A():
    x = 10
    def B():
        x = 20
        def C():
            global x
            x = x + 30
            print('3',x)
        C()
        print('2',x)
    
    B()
    print('1',x)
 
A()
```

## 클로저 사용

```python
def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b    # 함수 바깥쪽에 있는 지역 변수 a, b를 사용하여 계산
    return mul_add          # mul_add 함수를 반환
 
c = calc()
print(c(1), c(2), c(3), c(4), c(5))
print(list(map(c,[1,2,3,4])))
```

```python
# lambda로 클로저 만들기
def calc():
    a = 3 
    b = 5
    return lambda x : x * a + b 

c = calc()
print(c(1),c(2),c(3),c(4),c(5))
```

```python
#클로저에 지역변수 변경하기
def calc():
    a=3
    b=5
    total = 0

    def mul_add(x):
        nonlocal total
        total = total + a*x+b
        return total
    return mul_add

c= calc()
print(c(1),c(2),c(3),c(4),c(5))
```
