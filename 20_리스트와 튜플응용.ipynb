```python
# 리스트 컴프리핸션
## 1. [식 for 변수 in 리스트]
## 2. list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print(a)

for i in range(10):
    print(i)
```

```python
b = list(j for j in range(10))
print(b)
```

```python
c = [i+5 for i in range(10)]
print(c)
```

```python
d = [i*5 for i in range(10)]
print(d)
```

```python
# 리스트 표현해서 if 조건문 사용하기

## 1. [식 for 변수 in 리스트 if 조건식]
## 2. list(식 for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)

b = [a+5 for a in range(10) if a % 2 == 1]
print(b)
```

```python
# for 반복문과 if 조건문을 여러 번 사용하기
# [for 변수1 in 리스트1 if 조건식1
#  for 변수2 in 리스트2 if 조건식2]

a = [i * j for j in range(2,10)
     for i in range(1,10)]
print(a)
```

```python
# 리스트에 map 사용하기

## list(map(함수,리스트))
## tuple(map(함수,튜플))

a = [1.2, 2.5, 3.7, 4.6]
for i in range(len(a)):
    a[i] = int(a[i])

print(a)

a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int,a))
print(a)
```

```python
#두수 사이의 숫자중 3과 7로 끝나는 않은 숫자 출력하기
a , b = list(map(int,input().split()))
lis = []
for i in range(a,b+1):
    if i % 10 != 3 and i % 10 != 7:
        lis.append(i)
print(*lis)
```

```python
for i in range(a,b+1):
    if i % 10 == 3 or i % 10 == 7:
        continue
    print(i, end=" ")
    
```

```python
# 역삼각형
# *********
#  *******
#   *****
#    ***
#     *
    
n = int(input())

for i in range(n):
    for j in range(i):
        print(" ", end="")
    for j in range(2 * (n-i) - 1):
        print("*", end="")
    print()
```
