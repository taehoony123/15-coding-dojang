```python
# continue로 코드 실행 건너뛰기

for i in range(10):
    if i % 2 == 0:
        continue # 짝수면 건너뛰어서 print 문 실행
    print(i)
```

```python
#while 반복문에서 continue로 코드 실행 건너뛰기

i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```

```python
#  입력한 횟수 대로 반복하기
count = int(input("숫자"))

i = 1
while True :
    print(i)
    i += 1
    if i == count+1:
        break
```

```python
#입력한 숫자 까지 홀수 출력하기
count = int(input("숫자"))

for i in range(count + 1):
    if i % 2 == 0:
        continue
    print(i)
```

```python
total = int(input())
fee = 1350

while True:
    total -= fee
    if total > 0 :
        print(total)
    else:
        print("잔액 부족")
        break
    
```
