```python
# 변수 여러 개를 한 번에 만들기 

x, y, z = 10, 20, 30 
print(x)
print(y)
print(z)
```

```python
x = y = z = 10 
print(x)
print(y)
print(z)
```

```python
x, y = 10, 20 
x, y = y, x

print(x)
print(y)
```

```python
# 입력 값을 정수로 변환하기 
a = int(input("첫 번째 숫자 : "))
b = int(input("두 번째 숫자 : "))

print(a + b)
```

```python
# 두 숫자의 합 구하기 
a, b = input("숫자 2 개를 입력하세요: ").split()
print(a) 
print(b)
print(a + b)
```

```python
# 두 숫자의 합 구하기 
a, b = input("숫자 2 개를 입력하세요: ").split()
a = int(a)
b = int(b)
print(a + b)
```

```python
# 입력받은 값을 콤마를 기준으로 분리하기
a, b = map(int, input("숫자 두 개를 입력하세요: ").split())

# a, b = map(int, ["10", "20"])

# a, b = 10, 20 

print(a + b)
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요: ").split(","))

print(a + b)
```
