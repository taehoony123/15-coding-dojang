그다음에 피타고라스의 정리에서 c의 길이를 계산하려면 제곱근을 구해야 합니다.

그럼 √(루트)는 어떻게 구현해야 할까요? 이때는 math 모듈의 sqrt 함수를 사용하면 편리합니다. sqrt는 제곱근을 뜻하는 square root에서 따왔습니다.

math.sqrt(값)

```python
import math

class Point2D():
    def __init__(self,x,y):
        self.x = x
        self.y = y

p1 = Point2D(x=30 , y=20)
p2 = Point2D(x=60 , y=50)

a = p2.x - p1.x
b = p2.y - p1.y

c = math.sqrt((a*a) + (b*b))
print(c)
```
