## 함수에서 위치 인수와 키워드 인수 사용하기

### 위치 인수와 리스트 언패킹 사용

````python
#  위치 인수를 사용하는 함수를 만들고 호출하기
def print_num(a,b,c):
    print(a)
    print(b)
    print(c)

print_num(10,20,30)
````

````python
# 언패킹 사용하기
x = [10,20,30]
print_num(*x)
````

 리스트(튜플) 앞에 *를 붙이면 언패킹(unpacking)이 되어서 print_numbers(10, 20, 30)과 똑같은 동작이 됩니다. 말 그대로 리스트의 포장을 푼다는 뜻입니다.

## 가변 인수 함수 만들기

그럼 위치 인수와 리스트 언패킹은 어디에 사용할까요? 이 기능들은 인수의 개수가 정해지지 않은 가변 인수(variable argument)에 사용합니다. 즉, 같은 함수에 인수 한 개를 넣을 수도 있고, 열 개를 넣을 수도 있습니다. 또는, 인수를 넣지 않을 수도 있습니다.

다음과 같이 가변 인수 함수는 매개변수 앞에 *를 붙여서 만듭니다.

````python
def print_num(*args):
    for arg in args:
        print(arg)

````

````python
print_num(10,20,30,40)
````

````python
y = [10,20,30,40]

print_num(y)
print_num(*y)
````

## 키워드 인수 사용하기

````python
def personal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)
````

````python
personal_info('홍길동',20,"서울")
````

````python
personal_info(age=30, address='서울시 용산구 이촌동', name='홍길동')
````

## 키워드 인수와 딕셔너리 언패킹 사용

````python
x = {'name': '홍길동', 'age': 30, 'address': '서울시 용산구 이촌동'}
personal_info(**x)
````

위에랑 같음

````python
personal_info(**{'name': '홍길동', 'age': 30, 'address': '서울시 용산구 이촌동'})
````

personal_info에 *x를 넣으면 x의 키가 출력됩니다. 즉, 딕셔너리를 한 번 언패킹하면 키를 사용한다는 뜻이 됩니다

````python
personal_info(*x)
````

## 키워드 인수를 사용하는 가변 인수 함수 만들기

````python
def personal_info(**kwargs):
    for kw, arg in kwargs.items():
        print(kw, ': ', arg, sep='')
````

````python
personal_info(name='홍길동')
````

````python
personal_info(name='홍길동', age=30, address='서울시 용산구 이촌동')
````

````python
x = {'name':'홍길동'}
personal_info(**x)
````

````python
y = {'name': '홍길동', 'age': 30, 'address': '서울시 용산구 이촌동'}
personal_info(**y)
````

딕셔너리 x는 {'name': '홍길동'}이므로 <br>personal_info(**x)로 호출하면 personal_info(name='홍길동')과 같고,<br> 딕셔너리 y는 {'name': '홍길동', 'age': 30, 'address': '서울시 용산구 이촌동'}이므로 personal_info(name='홍길동', age=30, address='서울시 용산구 이촌동')과 같습니다.

## 연습문제

````python
korean, english, mathematics, science = 100, 86, 81, 91

def get_max_score(*args):
    return max(args)

max_score = get_max_score(korean, english, mathematics, science)
print('높은 점수:', max_score)
 
max_score = get_max_score(english, science)
print('높은 점수:', max_score)
````

## 심사문제

````python
korean, english, mathematics, science = map(int, input().split())

#가변 인자 -> 튜플로 처리
def get_min_max_score(*args): #korean, english, mathematics, science 등 여러 개 인자들을 args 변수에 담는다.
    return max(args),min(args)

#가변인자 키워드 -> 딕셔너리 처리
def get_average(**kwargs): #kwargs는 {'korean': 90, 'english': 80}으로 처리
    return sum(kwargs.values())/len(kwargs)

min_score, max_score = get_min_max_score(korean, english, mathematics, science)
average_score = get_average(korean=korean, english=english, mathematics=mathematics, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
 
min_score, max_score = get_min_max_score(english, science)
average_score = get_average(english=english, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'.format(min_score, max_score, average_score))
````

## 1. `*args` (가변 인자)
> **개수가 정해지지 않은 일반 값들을 묶어서 튜플(Tuple)로 처리할 때 사용합니다.**

* **개념:** 매개변수 앞에 별표 하나(`*`)를 붙이면, 함수를 호출할 때 넣은 여러 개의 인자들을 파이썬이 하나의 튜플로 묶어서 `args`라는 변수에 담아줍니다.
* **예시:** `get_min_max_score(10, 20, 30)` 형태로 호출하면, 함수 내부에서 `args`는 `(10, 20, 30)`이라는 튜플이 됩니다.

### 🔍 활용 메서드 & 키워드
* **`min(args)`, `max(args)`**
  파이썬 내장 함수인 `min()`과 `max()`를 사용해 묶여있는 점수들 중 가장 작은 값과 가장 큰 값을 찾습니다.
* **`return`**
  두 값을 콤마(`,`)로 연결해 반환하면 파이썬은 이를 튜플 형태로 감싸서 내보냅니다. 덕분에 함수 밖에서 다음과 같이 여러 변수로 한 번에 나누어 담는 **언패킹(Unpacking)**이 가능합니다.

## 2. `**kwargs` (가변 키워드 인자)
> **개수가 정해지지 않은 `이름=값` 쌍들을 묶어서 딕셔너리(Dictionary)로 처리할 때 사용합니다.**

* **개념:** 매개변수 앞에 별표 두 개(`**`)를 붙이면, `korean=90, english=80` 처럼 `키워드=값` 형태로 전달된 인자들을 하나의 딕셔너리로 묶어줍니다.
* **예시:** `get_average(korean=90, english=80)` 형태로 호출하면, 함수 내부에서 `kwargs`는 `{'korean': 90, 'english': 80}`이 됩니다.

### 🔍 활용 메서드 & 연산
* **`kwargs.values()`**
  딕셔너리에서 과목 이름(Key)은 빼고, 실제 점수들(Value)인 `[90, 80]`만 쏙 골라내는 메서드입니다.
* **`sum(...) / len(...)`**
  골라낸 점수들을 `sum()`으로 모두 더한 뒤, 딕셔너리에 들어있는 과목 개수인 `len(kwargs)`로 나누어 평균을 구합니다.
