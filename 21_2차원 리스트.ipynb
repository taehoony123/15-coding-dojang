```python
# for range 사용
a = [[10,20],
     [30,40],
     [50,60]]

for i in range(len(a)):
    for j in range(len(a[i])):
        print(a[i][j], end=" ")
    print()
```

```python
# sorted 2차원 리스트 정렬
students = [
    ['john', 'C', 19],
    ['maria', 'A', 25],
    ['andrew', 'B', 7]
]

print(sorted(students)) # 이름으로 정렬

print(sorted(students, key=lambda student : student[1])) # 성적순으로 정렬
print(sorted(students, key=lambda student : student[2])) # 나이순으로 정렬
```

```python
# 2차원 리스트의 할당과 복사 알아보기

# 리스트를 다른 변수에 할당해도 변수 이름만 달라질 뿐 같은 리스트(객체)를 참조한다.
a = [[10,20],
     [30,40],
    ]
b = a
b[0][0] = 500

print(a) # a도 변경됨
print(b)

# 주소가 같음
print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

```python
a = [[10,20],
     [30,40],
    ]
b = a.copy() # 얕은 복사
b[0][0] = 500

print(a) # a도 변경됨
print(b)

print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

```python
x = [1,2,3,4]
y = x.copy()
y[0] = 500

print(x)
print(y)

print(id(x))
print(id(y))
```

```python
import copy

a = [[10,20],
     [30,40],
    ]

b = copy.deepcopy(a) # 깊은 복사
b[0][0] = 500
print(a) # a는 변경되지 않음
print(b)

print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

```python
#표준 입력으로 2차원 리스트의 가로(col)와 세로(row)가 입력되고 그 다음 줄부터 리스트의 요소로 들어갈 문자가 입력됩니다. 이때 2차원 리스트 안에서 *는 지뢰이고 .은 지뢰가 아닙니다. 지뢰가 아닌 요소에는 인접한 지뢰의 개수를 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다).

#여러 줄을 입력 받으려면 다음과 같이 for 반복문에서 input을 호출한 뒤 append로 각 줄을 추가하면 됩니다(list 안에 문자열을 넣으면 문자열이 문자 리스트로 변환됩니다).

# 가로, 세로 입력 받기
col, row = map(int, input().split())

#지뢰 맵(2차원 리스트) 입력 받기
matrix = []
for i in range(row):
    matrix.append(list(input()))

# 8방향 탐색을 위한 dr, dc 리스트
dr = [-1, 1, 0, 0, -1, -1, 1, 1]
dc = [0, 0, -1, 1, -1, 1, -1, 1]

for r in range(row):
    for c in range(col):
        if matrix[r][c] == '*': # 지뢰인 경우
            continue
        
        count = 0 # 인접한 지뢰의 개수
        for i in range(8): # 8방향 탐색
            nr = r + dr[i]
            nc = c + dc[i]
            
            # 범위를 벗어나지 않고 지뢰인 경우
            if 0 <= nr < row and 0 <= nc < col and matrix[nr][nc] == '*':
                count += 1
        
        matrix[r][c] = str(count) # 지뢰가 아닌 요소에 인접한 지뢰의 개수 저장

# 결과 출력
for r in range(row):
    for c in range(col):
        print(matrix[r][c], end="")
    print()
```
