```python
# sorted 2차원 리스트 정렬
students = [
    ['john', 'C', 19],
    ['maria', 'A', 25],
    ['andrew', 'B', 7]
]

print(sorted(students)) # 이름으로 정렬

print(sorted(students, key=lambda student : student[1])) # 성적순으로 정렬
print(sorted(students, key=lambda student : student[2])) # 나이순으로 정렬
```

```python
# 2차원 리스트의 할당과 복사 알아보기

# 리스트를 다른 변수에 할당해도 변수 이름만 달라질 뿐 같은 리스트(객체)를 참조한다.
a = [[10,20],
     [30,40],
    ]
b = a
b[0][0] = 500

print(a) # a도 변경됨
print(b)
print(a is b) # a와 b는 같은 리스트를 참조합니다
# 주소가 같음
print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

```python
a = [[10,20],
     [30,40],
    ]
b = a.copy() # 얕은 복사
b[0][0] = 500

print(a) # a도 변경됨
print(b)

print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

```python
x = [1,2,3,4]
y = x.copy()
y[0] = 500

print(x)
print(y)
print(x is y)
print(id(x))
print(id(y))
```

```python
import copy

a = [[10,20],
     [30,40],
    ]

b = copy.deepcopy(a) # 깊은 복사
b[0][0] = 500
print(a) # a는 변경되지 않음
print(b)

print(id(a)) # a의 주소
print(id(b)) # b의 주소
```

## 지뢰찾기

```python
# 2차원 리스트의 가로(col)와 세로(row)가 입력
col, row = map(int, input().split())

# 2차원 리스트 만들기
matrix = []
for i in range(row):
    matrix.append(list(input()))

# 8방향 탐색을 위한 행과 열의 변화량
dr = [-1,1,0,0,-1,-1,1,1]
dc = [0,0,-1,1,-1,1,-1,1]

# 지뢰가 아닌 요소에 인접한 지뢰의 개수 저장하기
for r in range(row):
    for c in range(col):
        # 지뢰인 요소는 건너뛰기
        if matrix[r][c] == '*':
            continue

        count = 0
        # 8방향 탐색하기
        for i in range(8): # 0~7 8방향 탐색
            nr = r + dr[i]
            nc = c + dc[i]
            # 인접한 요소가 지뢰인지 확인하기
            if (0<= nr < row) and (0 <= nc < col) and matrix[nr][nc] == '*':
                count += 1
            
        # 지뢰가 아닌 요소에 인접한 지뢰의 개수 저장
        matrix[r][c] = str(count)

# 2차원 리스트 출력하기
for r in range(row):
    for c in range(col):
        print(matrix[r][c], end="")
    print()
```
