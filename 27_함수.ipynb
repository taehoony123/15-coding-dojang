## 함수에서 값을 여러 개 반환하기

```python
def add_sub(a,b):
    return a+b, a-b

add_sub(10,20)
```

```python
def mul(a,b):
    c = a*b
    return c

def add(a,b):
    c = a+b
    print(c)
    d = mul(a,b)
    print(d)

x = 10
y = 20

add(x,y)
```

## 심사문제

```python
x,y = map(int,input().split())

def calc(a,b):
    return a+b, a-b, a*b, a/b
    
a,s,m,d = calc(x,y)
print('덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}'.format(a, s, m, d))
```

## 과제

```python
scores = list(map(int, input().split()))

def get_min_max_score(*args): 
    return min(args),max(args)

def get_sum(*args):
    return sum(args)

def get_average(*args): 
    return sum(args)/len(args)

def get_variance(*args):
    mean_val = sum(args) / len(args)
    return sum((scores - mean_val) ** 2 for scores in args) / len(args)

min_score, max_score = get_min_max_score(*scores)
sum_score = get_sum(*scores)
average_score = get_average(*scores)
variance_score = get_variance(*scores)

print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 총점: {2:.2f} 평균 점수: {3:.2f} 분산 점수: {4:.2f}'.format(min_score, max_score,sum_score, average_score, variance_score))
 
```

## 과제2

```python
prices = {
    "아메리카노": 2500,
    "카페라떼": 3500,
    "바닐라라떼": 4000,
    "초코케이크": 5500,
    "샌드위치": 6000
}

def order(**kwargs):
    total_price = 0
    #kwargs.items는 prices안에 상품과 개수를 동시에 꺼냄
    for item,count in kwargs.items():
        if item in prices :
            item_price = prices[item] * count
            total_price += item_price
            print(f"{item}={count}개 : {item_price}원")
    
    print(f"총 결제 금액: {total_price}원")
    print("-" * 20)

order(아메리카노=2, 카페라떼=1)
order(바닐라라떼=2, 초코케이크=1, 샌드위치=1)
order(아메리카노=1, 샌드위치=2)
```
