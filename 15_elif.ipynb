```python
# elif 사용하기
x = 20

if x == 10:
    print("10입니다")
elif x == 20:
    print("20입니다")
```

```python
x = 30

if x == 10:
    print("10입니다")
elif x == 20:
    print("20입니다")
else:
    print("10도 아니고 20도 아닙니다")
```

```python
# 음료수 자판기
button = int(input())

if button == 1 :
    print("콜라")
elif button == 2 :
    print("사이다")
elif button == 3 :
    print("탄산수")
else:
    print("재고하지 않는 메뉴이다")
```
