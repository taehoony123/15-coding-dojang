```python
# 왼쪽 공백 삭제하기
print("   hello   ".lstrip())

# 왼쪽 특정 문자 삭제하기
print("www.example.com".lstrip("w"))
```

```python
# 오른쪽 공백 삭제하기
print("   hello   ".rstrip())

# 오른쪽 특정 문자 삭제하기
print("www.example.com".rstrip(".com"))
```

```python
# 양쪽 공백 삭제하기
print("   hello   ".strip())

# 양쪽 특정 문자 삭제하기
print("www.example.com".strip("w.com"))
```

```python
# 문자열 왼쪽 정렬
print("hello".ljust(10))

# 문자열 오른쪽 정렬
print("hello".rjust(10))

# 문자열 가운데 정렬
print("hello".center(10))
```

```python
# 메서드 체이닝
# 메서드 체이닝은 메서드를 줄줄이 연결한다고 해서 메서드 체이닝(method chaining)

# 다음은 문자열을 대문자로 바꾼 뒤 왼쪽 정렬하는 예입니다.
print("hello".upper().ljust(10))
# 다음은 문자열을 오른쪽 정렬한 뒤 양쪽 공백을 삭제하는 예입니다.
print("   hello   ".rjust(10).strip())
```

```python
# 문자열 왼쪽에 0 채우기
print("123".zfill(10))
```

```python
# 문자열 위치 찾기
# find =  특정 문자열을 찾아서 인덱스를 반환하고, 문자열이 없으면 -1을 반환
print("apple pineapple".find("pl"))

# 문자열 위치 찾기(오른쪽부터)
print("apple pineapple".rfind("pl"))

# index = 특정 문자열을 찾아서 인덱스를 반환하고, 문자열이 없으면 오류 발생
print("aaaaa pineapple".index("pl"))

# 문자열 위치 찾기
print("apple pineapple".index("pl"))  # 2
# 오른쪽 부터 문자열 위치 찾기
print("apple pineapple".rindex("pl")) # 12

# 문자열 개수 세기
print("apple pineapple".count("pl")) # 2
```

## Formating

```python
# %s
print("I have a %s." % "apple")

name = "Alice"
print("My name is %s." % name)
```

```python
# %d
print("i am %d years old." % 20)
```

```python
# %f (소수점 이하 6자리까지)
print("pi is %f" % 3.141592653589793)

# %.3f (소수점 이하 3자리까지)
print("pi is %.3f" % 3.141592653589793)

# %.2f (소수점 이하 2자리까지)
print("pi is %.2f" % 3.141592653589793)
```

```python
# 서식 지정자로 문자열 정렬하기

# % 뒤에 숫자를 붙이면 문자열을 지정된 길이로 만든 뒤 오른쪽으로 정렬하고 남는 공간을 공백으로 채웁니다.

print("%10s" % "hello") # 오른쪽 정렬
print("%-10s" % "hello") # 왼쪽 정렬
```

```python
# 자릿수가 다른 숫자 출력하기
print("%d" % 42) # 42
print("%5d" % 42) #    42

# f
print("%10.3f" % 3.141592653589793) # 3.141593
print("%.3f" % 3.141592653589793) # 3.14
```

```python
# 서식지정자로 문자열 안에 값 여러개 넣기
print("I have a %s and a %s." % ("apple", "banana"))
```

## format method 사용하기

```python
# {}
print("I have a {} and a {}.".format("apple", "banana"))
```

```python
# 여러개 넣기
print("I have a {0} and a {1}.".format("apple", "banana"))
```

```python
# 인덱스 생략 - 순서대로 들어감
print("I have a {} and a {}.".format("apple", "banana"))
```

```python
#  인덱스 대신 이름 정하기
print("I have a {fruit1} and a {fruit2}.".format(fruit1="apple", fruit2="banana"))
```

```python
# 문자열 포매팅에 변수를 그대로 사용
fruit1 = "apple"
fruit2 = "banana"
print("I have a {0} and a {1}.".format(fruit1, fruit2))
```

```python
# 문자열 정렬
print("{0:<10}".format("apple")) # 왼쪽 정렬
print("{0:>10}".format("apple")) # 오른쪽 정렬
```

```python
a = input("문자열을 입력하세요: ")

words = a.split()

count = 0
for word in words:
    cleaned_word = word.strip('.,!?";()')  # 구두점 제거

    if cleaned_word == "the":  # 빈 문자열이 아닌 경우에만 카운트
        count += 1
print(count)
```

```python
text = input()

texts = text.replace('.', ' ').replace(',', ' ').replace('?', ' ').replace('!', ' ').replace(';', ' ').replace('"', ' ').replace('(', ' ').replace("'", ' ' )
print(texts)

words = texts.split()

count = words.count('the')

print(count)
```

```python
price = (input())

price_string = price.replace(';', ' ')

prices = []
for price in price_string.split():
    prices.append(int(price))

prices.sort(reverse=True)

for price in prices:
    print(f"{price:,}".rjust(10))
```
