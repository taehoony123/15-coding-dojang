```python
x = {
    'a' : 10,
    'b' : 20,
    'c' : 30,
    'd' : 40
}

x.setdefault('e') # 키,값 쌍 추가
x.update({'f': 50}) # 키의 값 수정, 키가 없으면 키-값 쌍 추가
x.update(a = 5) # 키의 값 수정, 키가 없으면 키-값 쌍 추가
x.update(a=900, f=60) # a의 값 수정, f의 값 수정
x.setdefault('g', 60) # 키,값 쌍 추가
x.setdefault('a', 100) # 이미 존재하는 키이므로 아무 일도 일어나지 않음
x
```

```python
y = {
    1: 'one',
    3: 'three',
}

y.update({2: 'two'}) # 키,값 쌍 추가


print(y)
```

```python
y.update([[2,'two'], [4, 'four']]) # 키,값 쌍 추가
y
```

```python
y.update(zip([1,2],['one', 'two'])) # 키,값 쌍 추가
y
```

## 딕셔너리 키-값 삭제하기

```python
# 딕셔너리 키값 쌍 삭제하기
print(y.pop(4))
print(y)
print(y.pop(4, 'return')) # 뒤에 'return'은 4라는 키가 없을 때 반환할 값입니다.
print(y)
```

## 값 불러오기

```python
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}
print(x.get('a')) # 키 'a'에 대응하는 값 반환
print(x.get('e')) # 키 'e'는 존재하지 않으므로 None 반환
print(x.items()) # 딕셔너리의 모든 키-값 쌍을 (키, 값) 튜플로 묶은 뷰 객체 반환
print(x.keys()) # 딕셔너리의 모든 키를 담은 뷰 객체 반환
print(x.values()) # 딕셔너리의 모든 값을 담은 뷰 객체
```

## 리스트와 튜플로 딕셔너리 만들기

```python
keys = ['a', 'b', 'c', 'd']
#dict.fromkeys는 키 리스트로 딕셔너리를 생성하며 값은 모두 None으로 저장
x = dict.fromkeys(keys) # keys 리스트의 요소를 키로 하고, None을 값으로 하는 딕셔너리 생성
print(x)

x = dict.fromkeys(keys, 100) # keys 리스트의 요소를 키로 하고, 100을 값으로 하는 딕셔너리 생성
print(x)
```

```python
from collections import defaultdict
#defaultdict 사용하기
#지금까지 사용한 딕셔너리(dict)는 없는 키에 접근했을 경우 에러가 발생합니다

x = {'a': 0 , 'b': 0, 'c': 0, 'd': 0}
# x['z'] # 에러
y = defaultdict(int) # int는 기본값이 0인 자료형입니다.
y['z'] # 0 반환
print(y)


z = defaultdict(lambda: 'python') # lambda는 익명 함수를 만드는 키워드입니다.
print(z['a'])
print(z[0])
```
