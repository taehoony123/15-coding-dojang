## 리스트와 튜플로 딕셔너리 만들기

```python
keys = ['a', 'b', 'c', 'd']
#dict.fromkeys는 키 리스트로 딕셔너리를 생성하며 값은 모두 None으로 저장
x = dict.fromkeys(keys) # keys 리스트의 요소를 키로 하고, None을 값으로 하는 딕셔너리 생성
print(x)

x = dict.fromkeys(keys, 100) # keys 리스트의 요소를 키로 하고, 100을 값으로 하는 딕셔너리 생성
print(x)
```

```python
from collections import defaultdict
#defaultdict 사용하기
#지금까지 사용한 딕셔너리(dict)는 없는 키에 접근했을 경우 에러가 발생합니다

x = {'a': 0 , 'b': 0, 'c': 0, 'd': 0}
# x['z'] # 에러
y = defaultdict(int) # int는 기본값이 0인 자료형입니다.
y['z'] # 0 반환
print(y)


z = defaultdict(lambda: 'python') # lambda는 익명 함수를 만드는 키워드입니다.
print(z['a'])
print(z[0])
```

## 딕셔너리 표현식 사용하기

```python
keys = ['a', 'b', 'c', 'd']
 # dict.fromkeys로 키 리스트와 값으로 딕셔너리를 생성한 후, 딕셔너리 컴프리헨션으로 키-값 쌍을 다시 딕셔너리로 생성
x = {key: value for key, value in dict.fromkeys(keys).items()}
print(x)
```

```python
keys = [key for key in dict.fromkeys(['a', 'b', 'c', 'd']).keys()]
print(keys)

values = [value for value in {'a': 10, 'b': 20, 'c': 30, 'd': 40}.values()]
print(values)
```

```python
# 키-값 자리를 바꿈
changes = {value: key for key, value in {'a': 10, 'b': 20, 'c': 30, 'd': 40}.items()}  
print(changes)
```

## 딕셔너리 if 조건문 사용

```python
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

for key , value in x.items():
    if value >= 30:
        print(key, value)
```

```python
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

for key , value in x.items():
    if value == 20:
        del x[key] # 딕셔너리에서 키-값 쌍 삭제
print(x)


#별 문제 없이 잘 삭제될 것 같지만 반복 도중에 딕셔너리의 크기가 바뀌었다는 에러가 발생합니다. 
#즉, 딕셔너리는 for 반복문으로 반복하면서 키-값 쌍을 삭제하면 안 됩니다.
```

```python
for key , value in list(x.items()):
    if value == 20:
        del x[key] # 딕셔너리에서 키-값 쌍 삭제

print(x)
```

```python
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

x = {key: value for key, value in x.items() if value != 20} # 딕셔너리 컴프리헨션으로 키-값 쌍을 새로 만들어서 삭제
print(x)
```

## 딕셔너리 할당

```python
x = {
    'a': 0,
    'b': 0,
    'c': 0,
    'd': 0
}

y = x # x와 y는 같은 딕셔너리를 참조합니다.
print(x)
print(y)
print(x is y)

print()

y['a'] = 100 # y의 키 'a'에 대응하는 값을 100으로 수정
print(x) # x도 변경됩니다.
print(y)
```

```python
x = {'a' : 0, 'b' : 0, 'c' : 0, 'd' : 0}

y= x.copy() # x의 얕은 복사본을 생성하여 y에 저장
print(x)
print(y)
print(x is y) # x와 y는 다른 딕셔너리를 참조합니다
print(x == y) # x와 y는 같은 키-값 쌍을 가지고 있습니다. 따라서 x와 y는 동등합니다.
```

##  중첩 딕셔너리의 할당과 복사 알아보기

```python
x = {'a': {'python': '2.7'}, 'b': {'python': '3.6'}}
print(x)

y = x.copy() # x의 얕은 복사본을 생성하여 y에 저장
print(y)

print(x is y) # x와 y는 다른 딕셔너리를 참조합니다


y['a']['python'] = '2.7.15'
print(x) # x도 변경됨
print(y)
print(x is y) # x와 y는 다른 딕셔너리를 참조합니다
print(x == y) # x와 y는 같은 키-값 쌍을 가지고 있습니다. 따라서 x와 y는 동등합니다.
```

```python
import copy
x = {'a': {'python': '2.7'}, 'b': {'python': '3.6'}}
y = copy.deepcopy(x) # x의 깊은 복사본을 생성하여 y에 저장
y['a']['python'] = '2.7.15' # y의 키 'a'에 대응하는 값인 딕셔너리에서 키 'python'에 대응하는 값을 '2.7.15'로 수정
print(x)
print(y)
print(x is y) # x와 y는 다른 딕셔너리를 참조합니다
print(x == y) # x와 y는 같은 키-값 쌍을 가지고
```

```python
import copy
x = {'python': {'version': '2.7'}, 'script': {'name': 'hello.py'}}
a = x
b = x.copy()
c = copy.deepcopy(x)
x['python']['version'] = '3.6'
print(a['python']['version'], b['python']['version'], c['python']['version'])
```
