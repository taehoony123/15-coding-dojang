```python
# 공백 찍기
n = 5
for i in range(n):
    for j in range(n):
        if j < i:
            print(" ", end= "")
        else :
            print("*", end="")
    print()
```

```python
n = int(input())

for i in range(n):  # i는 0부터 4까지 (층수)
    for j in range(2 * n - 1):  # j는 가로 칸 (0부터 8까지)
        # 핵심 조건: i층일 때 가운데를 중심으로 양옆으로 i만큼만 별을 찍는다!
        if (n - 1) - i <= j <= (n - 1) + i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
