## 세트 만들기 

```python
fruits = {'strawberry', 'grape','orange','pineapple','cherry'}
print(fruits)
```

```python
# 세트는 요소의 순서가 정해져 있지 않습니다(unordered).
# 따라서 세트를 출력해보면 매번 요소의 순서가 다르게 나옵니다.
fruits = {'strawberry', 'grape','orange','pineapple','cherry'}
print(fruits[0])  # 오류 발생
```

```python
# 세트에 들어가는 요소는 중복될 수 없습니다. 다음과 같이 세트에 'orange'를 두 개 넣어도 실제로는 한 개만 들어갑니다.
fruits = {'strawberry', 'grape','orange','pineapple','cherry','orange'}
print(fruits)
```

### 세트에 특정 값이 있는지 확인

```python
fruits = {'strawberry', 'grape','orange','pineapple','cherry'}
print('orange' in fruits)  # 'orange'가 세트에 있는지 확인
print('watermelon' not in fruits)  # 'watermelon'이 세트에 있는지 확인
```

### set을 사용하여 세트 만들기

```python
a = set('apple')
print(a)
```

```python
# range를 사용하면 0부터 4까지 숫자를 가진 세트를 만들 수 있습니다.
b = set(range(5))
print(b)
```

```python
# 빈 세트는 c = set()과 같이 set에 아무것도 지정하지 않으면 됩니다.
c = set()
print(c)
```

```python
c = {}
print(type(c))  # c는 빈 딕셔너리입니다. 세트가 아닙니다.
c = set()
print(type(c))  # c는 빈 세트입니다.
```

```python
# 세트는 세트 안에 세트를 넣을 수 없습니다. 
a = {{1, 2}, {3, 4}}
print(a)
```

```python
# 파이썬은 내용을 변경할 수 없는 세트도 제공합니다.
a = frozenset(range(10))
print(a)
```

```python
# frozenset은 집합 연산과 메서드에서 요소를 추가하거나 삭제하는 연산, 메서드는 사용할 수 없습니다
a = frozenset(range(10))

a.add(10)  # 오류 발생
a |= 10  # 오류 발생
a.remove(0)  # 오류 발생
```

### 집합 연산 사용하기

```python
# 집합연산
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)  # 합집합 
print(a & b)  # 교집합
print(a - b)  # 차집합
print(a ^ b)  # 대칭차집합
```

```python
print(set.union(a, b))  # a | b와 같습니다.
print(set.intersection(a, b))  # a & b와 같습니다.
print(set.difference(a, b))  # a - b와 같습니다.
print(set.symmetric_difference(a, b))  # a ^ b와 같습니다.
```

## 집합 연산 후 할당 연산자 사용하기

```python
a = {1, 2, 3, 4}

a |= {5, 6}  # a에 {5, 6}을 추가합니다. a = a | {5, 6}과 같습니다.
print(a)

a.update({7, 8})  # a에 {7, 8}을 추가합니다. a = a | {7, 8}과 같습니다.
print(a)

```

```python
a = {1, 2, 3, 4}

a &= {0,1,2,3,4}  # a에서 {0, 1, 2, 3, 4}과 겹치는 부분만 남깁니다. a = a & {0, 1, 2, 3, 4}과 같습니다.
print(a)

a.intersection_update({0,1,2,3,4})  # a에서 {0, 1, 2, 3, 4}과 겹치는 부분만 남깁니다. a = a & {0, 1, 2, 3, 4}과 같습니다.
print(a)
```

```python
a = {1, 2, 3, 4}

a -= {3}  # a에서 {2, 3}과 겹치는 부분을 제거합니다. a = a - {2, 3}과 같습니다.
print(a)

a.difference_update({3})  # a에서 {2, 3}과 겹치는 부분을 제거합니다. a = a - {2, 3}과 같습니다.
print(a)
```

```python
a = {1, 2, 3, 4}

print(a <= {1, 2, 3, 4})  # a가 {1, 2, 3, 4}의 부분집합인지 확인합니다. True

print(a.issubset({1, 2, 3, 4}))  # a가 {1, 2, 3, 4}의 부분집합인지 확인합니다. True

print(a >= {1,2,3,4,5})
print(a.issuperset({1,2,3,4,5}))
```

```python
a = {1,2,3,4}

print(a ^ {1,2,3})
```

## 세트가 겹치는지 확인

```python
a = {1,2,3,4}
print(a.isdisjoint({5,6,7,8}))  # a와 {5, 6, 7, 8}이 서로소인지 확인합니다. True
print(a.isdisjoint({3,4,5,6}))  # a와 {3, 4, 5, 6}이 서로소인지 확인합니다. False
```

## 세트 요소 활용

```python
a = {1, 2, 3, 4}
# 요소 추가
a.add(5)
print(a)
# 특정 요소 삭제
a.remove(2)
print(a)
a.discard(3)  # remove와 달리 요소가 없어도 오류가 발생하지 않습니다.
print(a)
# 임의의 요소 삭제
a.pop()
print(a)
# 모든 요소 삭제
a.clear()
print(a)
# 요소 개수 구하기
a = {1, 2, 3, 4}
print(len(a))
```

## 세트 할당과 복사

```python
a = {1,2,3,4}
b = a

print(a is b)
b.add(5)
print(b)
print(a)
```

```python
a = {1,2,3,4}
b = a.copy()
b.add(5)

print(a is b)
print(a == b)
```

## 공약수 구하기

1단계: 최대공약수 찾기

두 수를 동시에 나눌 수 있는 가장 큰 수를 구합니다. 

36과 60은 둘 다 12로 나누어 떨어집니다 

따라서 두 수의 최대공약수는 12입니다.

2단계: 최대공약수의 약수 구하기

1단계에서 구한 최대공약수 '12'의 약수를 모두 찾습니다. 

12의 약수: 1, 2, 3, 4, 6, 12

3단계: 공약수 확정

2단계에서 구한 약수들이 곧 36과 60의 공약수가 됩니다. 

정답: 1, 2, 3, 4, 6, 12


```python
n1, n2 = map(int, input().split())
a = {i for i in range(1, n1 + 1) if n1 % i == 0}
b = {i for i in range(1, n2 + 1) if n2 % i == 0}

divisor = a & b

result = 0

if type(divisor) == set:
    result = sum(divisor)

print(result)

# 36 60
```
